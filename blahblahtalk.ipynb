```python
import whisper
import gradio as gr
import openai
from TTS.api import TTS
```

```python
model_name = TTS.list_models()[9]
tts = TTS(model_name)
```

```python
tts.tts_to_file(text="Hi, I am English learning assistant", file_path="output.wav")
```

```python
model = whisper.load_model("medium")
```

```python
openai.api_key = 'API-key'
```

```python
def voice_chat(user_voice):

    messages = [{"role": "user", "content": f"You are language learning assitant, you will help me to study English language, try to keep it natural conversation, no boring and long sentenses, say OK."},
                   {"role": "assistant", "content": f"OK"}]


    user_message = model.transcribe(user_voice)["text"]

    reply = user_message

    messages.append(
        {"role": "user", "content": user_message},
    )
    print(messages)

    chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", messages=messages
    )

    reply = chat.choices[0].message.content

    messages.append({"role": "assistant", "content": reply})

    tts.tts_to_file(text=reply, file_path="output.wav")
    return('output.wav')
```

```python

with gr.Blocks(css=".imgs{height:400px};") as demo:

    with gr.Column(scale=2, max_width=200):

      gr.Markdown("""# BlahBlah Talk""")
      img = gr.Image('/content/drive/MyDrive/blahblah talk/Andy.png', elem_classes="imgs").style(full_width=False)

      with gr.Column():
        voice_reply = gr.Audio('output.wav')
        inp = gr.inputs.Audio(source="microphone", type="filepath")
        greet_btn = gr.Button("Submit")
        greet_btn.click(fn=voice_chat, inputs=inp, outputs=[voice_reply]),

demo.launch(debug=True, share=True)
```
